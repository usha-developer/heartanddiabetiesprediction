```python
import pickle
import pandas as pd
from sklearn.preprocessing import Imputer
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
import sklearn
import numpy as np
```

```python
#reading the two datasets for heart and diabetes
dh=pd.read_csv("framingham.csv")
dd=pd.read_csv("diabetes.csv")
```

```python
#feature selection for heart
di=dh[["male","age","currentSmoker","cigsPerDay","prevalentStroke","prevalentHyp","diabetes","totChol","sysBP","diaBP","heartRate","glucose"]]
do=dh.iloc[:,-1]
#featuire selection for diabetes
ddi=dd.iloc[:,1:-1]
ddo=dd.iloc[:,-1]
```

```python
#thus heat module contains null values
from sklearn.preprocessing import Imputer
im=Imputer(missing_values="NaN",strategy="mean",axis=0)
di["cigsPerDay"]=im.fit_transform(di[["cigsPerDay"]])
di["totChol"]=im.fit_transform(di[["totChol"]])
di["heartRate"]=im.fit_transform(di[["heartRate"]])
di["glucose"]=im.fit_transform(di[["glucose"]])
```

```python
di.isnull().sum()
do.isnull().sum()
ddi.isnull().sum()
ddo.isnull().sum()
```

```python
#for heart dieases training set
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(di,do,test_size=0.2)

#for diabetes
x1_train,x1_test,y1_train,y1_test=train_test_split(ddi,ddo,test_size=0.2)
```

```python
#for heart diseases
from sklearn.linear_model import LogisticRegression
lg=LogisticRegression()
lg.fit(x_train,y_train)
lg.predict(x_test)
lg.score(x_test,y_test)


#for diabets

from sklearn.linear_model import LogisticRegression
lg1=LogisticRegression()
lg1.fit(x1_train,y1_train)
lg1.predict(x1_test)
lg1.score(x1_test,y1_test)
```

```python
with open("heart.pickle","wb") as f:
    pickle.dump(lg,f)
with open("diabetes.pickle","wb") as g:
    pickle.dump(lg1,g)
```
